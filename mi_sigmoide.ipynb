```python
# Proyecto Final
#Grupo 1SF133
# Yissel Martínez yissel.martinez@utp.ac.pa
# Jorge Mendoza  jorge.mendoza3@utp.ac.pa
import numpy as np
import matplotlib.pyplot as plt

def sigmoide(t):
    return 1 / (1 + np.exp(-t))

# Solicitar al usuario ingresar los valores del vector t
valores_t = input("Ingresa los valores del vector t separados por espacios: ")
valores_t = valores_t.split()
valores_t = [float(valor) for valor in valores_t]

# Convertir la lista de valores a un arreglo NumPy
t = np.array(valores_t)

# Calcular la función sigmoide para el vector t
resultado = sigmoide(t)

# Imprimir los valores de t y el resultado de la función sigmoide
for i in range(len(t)):
    print("\nResultado de la función sigmoide para t:", t[i]," = ", resultado[i])

# Crear un arreglo de valores en el intervalo [-4, 4]
intervalo = np.linspace(-4, 4)

# Calcular la función sigmoide para el intervalo
resultado_intervalo = sigmoide(intervalo)

# Graficar la función sigmoide
plt.plot(intervalo, resultado_intervalo)
plt.title("Gráfico de la función sigmoide")
plt.xlabel("t")
plt.ylabel("sigmoide(t)")
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoide_general(x, w0, w1):
    return 1 / (1 + np.exp(-(w0 + w1 * x)))

# Solicitar al usuario ingresar el tamaño del vector para w0 y w1
tamaño = int(input("Ingresa el tamaño del vector para w0 y w1: "))

# Solicitar al usuario ingresar los valores para w0
valores_w0 = []
for i in range(tamaño):
    valor = float(input(f"Ingrese el valor {i+1} para w0: "))
    valores_w0.append(valor)

# Solicitar al usuario ingresar los valores para w1
valores_w1 = []
for i in range(tamaño):
    valor = float(input(f"Ingrese el valor {i+1} para w1: "))
    valores_w1.append(valor)

# Solicitar al usuario ingresar el valor de x
x = float(input("Ingresa el valor de x: "))

# Crear un arreglo de valores en el intervalo [0, 40]
intervalo = np.linspace(0, 40, num=100)

# Graficar la función sigmoide y mostrar el resultado de aplicar la función para cada combinación de w0 y w1
for w0, w1 in zip(valores_w0, valores_w1):
    resultado_intervalo = sigmoide_general(intervalo, w0, w1)
    plt.plot(intervalo, resultado_intervalo, label=f"w0={w0}, w1={w1}")

    # Calcular el resultado de aplicar la función sigmoide para los valores ingresados de w0, w1 y x
    resultado = sigmoide_general(x, w0, w1)
    print(f"\nResultado de aplicar la función sigmoide para w0={w0}, w1={w1}, x={x}: {resultado}")

# Agregar leyendas y descripción del punto medio
plt.title("\nGráfico de la función sigmoide generalizada")
plt.xlabel("x")
plt.ylabel("sigmoide_general(x)")
plt.axhline(y=0.5, color='gray', linestyle='--', label="Punto medio (y=0.5)")
plt.legend()
plt.grid(True)
plt.show()
0






```
